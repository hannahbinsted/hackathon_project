```python
# Add your Hackathon code here

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/netflix_titles.csv')
```

```python
df
```

```python
#Ensure that the 'date_added' column is in datetime format.

df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
#Ensure that the 'release_year' column is in the correct numeric format.

df['release_year'] = pd.to_numeric(df['release_year'], errors='ignore')
```

```python
#Extract month and year from the ' date_added' column for time-based analysis.

df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
```

```python
#For numeric columns, fill in missing values with the mean or median.

df.fillna(df.mean(), inplace=True)
```

```python
#For categorical columns 'rating', 'duration' and 'country', fill missing values with the mode or a placeholder like "Unknown", “Not Rated”

df['rating'].fillna('Not Rated', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
#Standardize capitalization in categorical columns 'type' and 'rating' for consistency.

df['type'] = df['type'].str.title()
df['rating'] = df['rating'].str.title()
```

```python
#Check for and remove any duplicate entries.

df.drop_duplicates(inplace=True)
```

```python
#Saved copy of cleaned data

df.to_csv('cleaned_data.csv', index=False)
```

```python
#New Dataframe

df2 = pd.read_csv('data/cleaned_data.csv')
df2
```

```python
#Number of movies/tv shows added from each year

total = df2['release_year'].value_counts().sort_index()

total.plot(kind='bar', figsize=(12, 6), color='blue')

plt.title('Number of Releases From Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')

plt.show()
```

```python
#Number of movies/tv shows released from each year seperately

grouped_data = df2.groupby(['release_year', 'type']).size().unstack(fill_value=0)

grouped_data.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Number of Movies and Tv Shows Released From Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')

plt.legend(title='Release Type')
plt.show()
```

```python
#Number of releases each year from each country

grouped_data = df2.groupby(['release_year', 'country']).size().unstack(fill_value=0)

grouped_data.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Countries where Releases Are From By Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')

plt.legend(title='Country')
plt.show()
```

```python
#Number of movies/tv shows added each year

total = df2['year_added'].value_counts().sort_index()

total.plot(kind='bar', figsize=(12, 6), color='blue')

plt.title('Number of Releases Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')

plt.show()
```
